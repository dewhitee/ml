```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score # for model accuracy analyze
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

dataset = np.concatenate((x, np.transpose(np.array([y]))), axis=1)
n, m = dataset.shape
train, test = train_test_split(dataset, random_state=1)
x_test = test[:, :m-1]
y_test = test[:, m-1]
x_train = train[:, :m-1]
y_train = train[:, m-1]

# градиентный спуск
clf = LogisticRegression()
clf.fit(x_train, y_train)

#print(x_test[:5])
#print(y_test[:5])
#clf2 = DecisionTreeClassifier(criterion='gini', random_state=1, max_depth=2)
#clf2.fit(x_train, y_train)


predicted = clf.predict(x_test)
print('Confusion matrix:')
print(confusion_matrix(y_test, predicted))
print('Test Accuracy: ', clf.score(x_test, y_test))
print('Precision: ', sum(precision_score(y_test, predicted, average=None)/3))
print('Recall: ', sum(recall_score(y_test, predicted, average=None)/3))
print('F1-score: ', sum(f1_score(y_test, predicted, average=None)/3))

#fig, axes  =plt.subplots(nrows=1, ncols=1, figsize =(2,2), dpi=200)
#plot_tree(clf, filled=True)
#plt.show()
```

```python
import matplotlib.pyplot as plt
iterations = []
accuracies = [] 

for i in range(1, 51):
    clf2 = LogisticRegression(max_iter=i)
    clf2.fit(x_train, y_train)
    accuracies.append(clf2.score(x_test, y_test))
    iterations.append(i)

plt.plot(iterations, accuracies)
plt.xlabel('n iterations')
plt.ylabel('accuracy')
plt.grid()
plt.show()
```
