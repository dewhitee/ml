```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score # for model accuracy analyze
```

```python
iris = load_iris()
#print(iris)

x = iris.data
y = iris.target
```

```python
# Machine learning begins
clf = DecisionTreeClassifier(criterion='gini', random_state=1) # by default it is indetermined - to make it
# determined we can specify random_state parameter (random seed of a classifier)

# Making error by not splitting dataset into training and test data intentionally to show what happens if we do so
clf.fit(x, y)
```

```python
import numpy as np
print('Accuracy: ', clf.score(x, y)) # gives one as we do not split by train and test (it is an error)
print(clf.predict(np.array([[1.2, 1.9, 0.9, .5]]))) # 0 = setosa
y_predicted = clf.predict(x)
print('Accuracy: ', accuracy_score(y, y_predicted))
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(4, 4), dpi=400)
plot_tree(clf, filled=True)
plt.show()
```

```python
print('Depth: ', clf.get_depth())
print('Leafs: ', clf.get_n_leaves())
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
import numpy as np
import matplotlib.pyplot as plt

clf2 = DecisionTreeClassifier(criterion='gini', random_state=1, max_depth=2)
score = cross_val_score(clf2, x, y, cv=4)
print('Score: ', score) # shows 1 because all complex ones are in the training set - this do not mean something
print('Average score: ', sum(score)/len(score))
```

```python
dataset = np.concatenate((x, np.transpose(np.array([y]))), axis=1)
n, m = dataset.shape
#splitter = round(0.75*n)
#test = dataset[:splitter]
#train = dataset[splitter:]
train, test = train_test_split(dataset, random_state=1)
x_test = test[:, :m-1]
y_test = test[:, m-1]
x_train = train[:, :m-1]
y_train = train[:, m-1]
print(x_test[:5])
print(y_test[:5])
clf2 = DecisionTreeClassifier(criterion='gini', random_state=1, max_depth=2)
clf2.fit(x_train, y_train)
predicted = clf2.predict(x_test)
print('Confusion matrix:')
print(confusion_matrix(y_test, predicted))
print('Test Accuracy: ', clf2.score(x_test, y_test))
print('Precision: ', sum(precision_score(y_test, predicted, average=None)/3))
print('Recall: ', sum(recall_score(y_test, predicted, average=None)/3))
print('F1-score: ', sum(f1_score(y_test, predicted, average=None)/3))

fig, axes  =plt.subplots(nrows=1, ncols=1, figsize =(2,2), dpi=200)
plot_tree(clf2, filled=True)
plt.show()
```
